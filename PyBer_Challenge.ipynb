### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head(10)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_by_type = pyber_data_df.groupby("type").count()["ride_id"]
rides_by_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_type = city_data_df.groupby("type").sum()["driver_count"]
drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
fares_by_type = pyber_data_df.groupby("type").sum()["fare"]
fares_by_type
```

```python
#  4. Get the average fare per ride for each city type
avg_fare_per_ride_by_type = fares_by_type / rides_by_type
avg_fare_per_ride_by_type
```

```python
# 5. Get the average fare per driver for each city type 
avg_fare_per_driver_by_type = fares_by_type / drivers_by_type
avg_fare_per_driver_by_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame ({"Total Rides": rides_by_type,
                                "Total Drivers": drivers_by_type,
                                "Total Fares": fares_by_type, 
                                "Average Fare per Ride": avg_fare_per_ride_by_type,
                                "Average Fare per Driver": avg_fare_per_driver_by_type})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name (i.e. say ignore the index name using method)(even though I don't see one?)
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
#Show data/value type in each column of imputs
pyber_data_df.dtypes
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date
pyber_city_type_date_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]

pyber_city_type_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_city_type_date_df = pyber_city_type_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
# The pivot() function will return a reshaped DataFrame that is organized by given index and column values
pyber_city_by_date_pivot = pyber_city_type_date_df.pivot(index = "date", columns = "type", values = "fare")
pyber_city_by_date_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_city_by_date_subpivot = pyber_city_by_date_pivot.loc[(pyber_city_by_date_pivot.index.get_level_values("date") >= "2019-01-01") & (pyber_city_by_date_pivot.index.get_level_values("date") <= "2019-04-29")]                                                          
pyber_city_by_date_subpivot
# We will need to clean up NAN values, and potentially transform the text in the date column to date notation
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
#YESSS! Transforming the date values 
pyber_city_by_date_subpivot.index = pd.to_datetime(pyber_city_by_date_subpivot.index)

# 7. Check that the datatype for the index is datetime using df.info()
pyber_city_by_date_subpivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
W_pyber_city_by_date_subpivot= pyber_city_by_date_subpivot.resample("W").sum()
```

```python
# 8 VISUALIZATIONS. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.pyplot as plt
import time 
import matplotlib.dates as mdates
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')


weekly_pyber_chart = W_pyber_city_by_date_subpivot.plot(figsize = (25, 7))

weekly_pyber_chart.set_title("Total Fare by City Type")
weekly_pyber_chart.set_ylabel("Fare ($USD)")
weekly_pyber_chart.set_xlabel("")

weekly_pyber_chart.get_figure().savefig("Analysis/Annotated_Chart_Total_Fares_by_City.png")
```
