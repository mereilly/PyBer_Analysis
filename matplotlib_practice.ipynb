```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), switching from using MATLAB to the object-oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# If you're graphing time-series data or ordinal values (e.g., age ranges, salary ranges, or groups), a vertical bar chart is a good choice. 
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.(no marker or linewidth, instead:)
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# To create a horizontal chart, we use the plt.barh() function
plt.barh(x_axis, y_axis)
```

```python
#  switch the arguments, moving the data on opposite axes
plt.barh(y_axis, x_axis)
```

```python
# The gca() method means "get current axes" to invert the y-axis to have the months in ascending order
# The bar should convey an amount on the x-axisnot a date
# switch data to graph on the opposite axes if we need to, 

plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# change bars to the color magenta
plt.barh(x_axis, y_axis, color="magenta")
plt.gca().invert_yaxis()
```

```python
# add legend for boston
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

plt.legend()

# add x and y labels on axes
plt.xlabel("Fare ($)")
plt.ylabel("Months")

#add title
plt.title("PyBer Fare by Month")
```
